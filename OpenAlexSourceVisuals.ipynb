```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('2026-04-22-UMN-Source.csv', index_col=0)
```

```python
pd.set_option("display.max_columns", 55)
```

```python
df.head() 
```

```python
def Matching_columns(df, col1, col2):
    match = df[col1].equals(df[col2])
    return match
```

```python
# primary topic
Matching_columns(df, 'primary_topic-display_name', 'topics-display_name')
```

```python
# subfield
Matching_columns(df, 'primary_topic-subfield', 'topics-subfield-id')
```

```python
# field
Matching_columns(df, 'primary_topic-field', 'topics-field-id')
```

```python
# domain
Matching_columns(df, 'primary_topic-domain', 'topics-domain-id')
```

```python
df['ror'].isna().sum(), df['orcid'].isna().sum()
```

```python
(df['ror'].isna() & df['orcid'].isna()).sum()
```

```python
df['ror'].notna().groupby(df['publication_year']).sum().reset_index().sort_values(by=['ror'], ascending=False).reset_index(drop=True)
```

```python
topics_display_name_ror_perc = df['ror'].count()
print('the # of ROR IDs assigned to DOIs is:', topics_display_name_ror_perc), print('the % of ROR IDs assigned to DOIs is:', round(topics_display_name_ror_perc / len(df) * 100, 3),'%')
```

```python
topics_display_name_ror = df['ror'].isna().groupby(df['topics-display_name']).sum().reset_index().sort_values(by=['ror'], ascending=False)
topics_display_name_ror['ror %'] = round(topics_display_name_ror['ror'] / sum(topics_display_name_ror['ror']) * 100, 4)
topics_display_name_ror.head(15).reset_index(drop=True)
```

```python
df['orcid'].notna().groupby(df['publication_year']).sum().reset_index().sort_values(by=['orcid'], ascending=False).reset_index(drop=True)
```

```python
topics_display_name_orcid_perc = df['orcid'].count()
print('the # of ORCID IDs assigned to DOIs is:', topics_display_name_orcid_perc), print('the % of ORCID IDs assigned to DOIs is:', round(topics_display_name_orcid_perc / len(df) * 100, 3),'%')
```

```python
topics_display_name_orcid = df['orcid'].isna().groupby(df['topics-display_name']).sum().reset_index().sort_values(by=['orcid'], ascending=False)
topics_display_name_orcid['orcid %'] = round(topics_display_name_orcid['orcid'] / sum(topics_display_name_orcid['orcid']) * 100, 4)
topics_display_name_orcid.head(15).reset_index(drop=True)
```

```python
topics_subfield_display_name_ror = df['ror'].isna().groupby(df['topics-subfield-display_name']).sum().reset_index().sort_values(by=['ror'], ascending=False)
topics_subfield_display_name_ror.head(15).reset_index(drop=True)
```

```python

topics_field_display_name_ror = df['ror'].isna().groupby(df['topics-field-display_name']).sum().reset_index().sort_values(by=['ror'], ascending=False)
topics_field_display_name_ror.head(15).reset_index(drop=True)
```
